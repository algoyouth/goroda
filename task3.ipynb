```python
import ipywidgets as widgets

slider = widgets.IntSlider(min=0, max=100, step=1, value=50, description='Целое число:')
display(slider)
```

```python
a = widgets.IntSlider(2)
b = widgets.IntSlider(2)
display(widgets.HBox([a, b]))
```

```python
print(a.value * b.value)
```

```python
button = widgets.Button(description="SUM")
output = widgets.Output()

display(button, output)

def on_button_clicked(button):
  with output:
      print(a.value + b.value)

button.on_click(on_button_clicked)
```

```python
out = widgets.Output(layout={'border': '1px solid black'})
out
```

```python
with out:
    for i in range(3):
        print(i, 'Hello world!')
```

```python
with out:
    display(slider)
```

```python
a = widgets.IntSlider(description='a')
b = widgets.IntSlider(description='b')
c = widgets.IntSlider(description='c')
def f(a, b, c):
    print('{}*{}*{}={}'.format(a, b, c, a*b*c))

out = widgets.interactive_output(f, {'a': a, 'b': b, 'c': c})

widgets.HBox([widgets.VBox([a, b, c]), out])
```

```python
def reverse(w):
  if w:
    return w[-1] + reverse(w[:-1])
  else:
    return ''
def f(w):
  print(reverse(w))
out = widgets.interact(f, w = 'Hello')
```

```python
from ipywidgets import interact
@interact(w="А роза упала на лапу Азора")
def reverse(w):
  if w:
    return w[-1] + reverse(w[:-1])
  else:
    return ''
```

```python
from ipywidgets import interact_manual
@interact_manual(word="А роза упала на лапу Азора")
def reverse(word):
  if word:
    return word[-1].upper() + reverse(word[:-1])
  else:
    return ''
```
