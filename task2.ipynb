# Наибольший общий делитель (алгоритм Евклида)

Находит НОД двух чисел рекурсивно (алгоритм Евклида)
* `НОД(a, b) = НОД(b, a mod b) если b != 0`
* `          = a если b == 0`

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
print("ввести А")
a = input()
print("ввести B")
b = input()
gcd_value = gcd(int(a), int(b))
print(f"НОД({a},{b}) = {gcd_value}")
```

```python
def gcd2(a, b):
    if (b != 0):
        return gcd2(b, a % b)
    else:
        return a
```

```python
print("ввести А")
a = input()
print("ввести B")
b = input()
gcd_value = gcd2(int(a), int(b))
print(f"НОД({a},{b}) = {gcd_value}")
```
