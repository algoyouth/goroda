```python
# Создание словаря
student = {
    "имя": "Анна",
    "возраст": 16,
    "класс": "10А",
    "предметы": ["математика", "физика", "информатика"]
}

print("Студент:", student)
print("Имя студента:", student["имя"])
print("Возраст:", student.get("возраст", "Неизвестно"))
```

```python
# Добавление и изменение
student["оценки"] = {"математика": 5, "физика": 4, "информатика": 5}
student["возраст"] = 17
print("Обновленная информация:", student)
```

```python
# Методы словарей
print("Ключи:", list(student.keys()))
print("Значения:", list(student.values()))
print("Пары ключ-значение:", list(student.items()))
```

```python
# Подсчет частоты элементов
text = "hello world"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print("Частота символов:", char_count)
```

```python
# То же самое с методом get
char_count2 = {}
for char in text:
    char_count2[char] = char_count2.get(char, 0) + 1
print("Частота символов (метод 2):", char_count2)
```

```python
# Создание множеств
fruits1 = {"яблоко", "банан", "апельсин"}
fruits2 = {"банан", "киви", "апельсин", "груша"}

print("Множество 1:", fruits1)
print("Множество 2:", fruits2)

# Операции с множествами
print("Объединение:", fruits1 | fruits2)
print("Пересечение:", fruits1 & fruits2)
print("Разность:", fruits1 - fruits2)
print("Симметрическая разность:", fruits1 ^ fruits2)

# Проверка принадлежности
print("'яблоко' в множестве 1:", "яблоко" in fruits1)

# Удаление дубликатов из списка
numbers_with_duplicates = [1, 2, 3, 2, 4, 1, 5, 3]
unique_numbers = list(set(numbers_with_duplicates))
print("Исходный список:", numbers_with_duplicates)
print("Уникальные элементы:", unique_numbers)

# Добавление и удаление элементов
fruits1.add("манго")
fruits1.discard("банан")  # Удалит, если есть
print("После изменений:", fruits1)
```
